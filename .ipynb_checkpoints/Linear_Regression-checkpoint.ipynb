```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
insurance = pd.read_csv("insurance.csv")
```

```python
# β₁ = Σ[(xᵢ - x̄)(yᵢ - ȳ)] / Σ[(xᵢ - x̄)²]
def calculateSlope(x, y, mx, my):
    cross_deviation = []
    squared_deviation = []
    i = 0
    
    while i < len(x):
        cross_deviation.append((x[i]-mx)*(y[i]-my))
        squared_deviation.append((x[i]-mx)**2) 
        i+=1

    sum_of_cross_deviation = sum(cross_deviation)
    sum_of_squared_deviation = sum(squared_deviation)

    beta_one = sum_of_cross_deviation / sum_of_squared_deviation
    
    return beta_one

#β₀ = ȳ - β₁x̄
def calculateIntercept(beta_one, mx, my):
    beta_zero = my - (beta_one * mx)
    return beta_zero

#y = β₀ + β₁x + ε
def calulateLinearRegression(x, y):
    if len(x) != len(y):
        print("Cols are not equal")
        return
    
    mx = x.mean()
    my = y.mean()
    
    beta_one = calculateSlope(x, y, mx, my)
    beta_zero = calculateIntercept(beta_one, mx, my)

    print(f"y = {beta_zero} + {beta_one} * x")
    return beta_zero, beta_one
```

```python
x = insurance['bmi']
y = insurance['charges']

beta_zero, beta_one = calulateLinearRegression(x, y)
```

```python
y_pred = [beta_zero + beta_one * xi for xi in x]

plt.scatter(x, y, label='Actual data')

plt.plot(x, y_pred, color='orange', label='Regression line')

# Labels and title
plt.xlabel(x.name)
plt.ylabel(y.name)
plt.title('Simple Linear Regression')

# Show plot
plt.show()
```
